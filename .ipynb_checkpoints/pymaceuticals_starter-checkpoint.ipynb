### Analysis
* Overall, it is clear that Capomulin is a viable drug regimen to reduce tumor growth.
* Capomulin had the most number of mice complete the study, with the exception of Remicane, all other regimens observed a number of mice deaths across the duration of the study. 
* There is a strong correlation between mouse weight and tumor volume, indicating that mouse weight may be contributing to the effectiveness of any drug regimen.
* There was one potential outlier within the Infubinol regimen. While most mice showed tumor volume increase, there was one mouse that had a reduction in tumor growth in the study. 

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
from scipy.stats import linregress


# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single dataset
merged_data=pd.merge(study_results, mouse_metadata, how="left", on="Mouse ID" )
merged_data.head()
```

```python
import numpy as np
```

## Summary Statistics

```python
drug_regimen=merged_data.set_index("Drug Regimen")
drug_regimen_df=pd.DataFrame(drug_regimen)
drug_regimen_df=drug_regimen_df.groupby(["Drug Regimen"])
drug_regimen_df.head()
```

```python
mean_tumor_volume=drug_regimen_df["Tumor Volume (mm3)"].mean()
mean_tumor_volume
```

```python
summary_statistics_df=pd.DataFrame({"Mean Tumor Volume": mean_tumor_volume})
summary_statistics_df
```

```python
median_tumor_volume=drug_regimen_df["Tumor Volume (mm3)"].median()
summary_statistics_df["Median Tumor Volume"]=median_tumor_volume
summary_statistics_df
```

```python
tumor_volume_var=drug_regimen_df["Tumor Volume (mm3)"].var()
summary_statistics_df["Tumor Volume Variance"]=tumor_volume_var
summary_statistics_df
```

```python
tumor_volume_std=drug_regimen_df["Tumor Volume (mm3)"].std()
summary_statistics_df["Tumor Volume Std. Dev."]=tumor_volume_std
summary_statistics_df
```

```python
tumor_volume_sem=drug_regimen_df["Tumor Volume (mm3)"].sem()
summary_statistics_df["Tumor Volume Std. Err."]=tumor_volume_sem
summary_statistics_df
```

## Quartiles, Outliers and Boxplots

```python
mice_df = pd.DataFrame(merged_data)
```

```python
# Calculate the final tumor volume of each mouse. 
# Start by getting the last (greatest) timepoint for each mouse
max_tumor = mice_df.groupby(["Mouse ID"]).max()
max_tumor = max_tumor.reset_index()
max_tumor
```

```python
# Merge this group df with the original dataframe to get the tumor volume at the last timepoint
# Calculate the final tumor volume of each mouse in Capomulin treatment regime
merged_data2 = max_tumor[['Mouse ID','Timepoint']].merge(mice_df,on=['Mouse ID','Timepoint'],how="left")
capomulin = merged_data2.loc[merged_data2["Drug Regimen"] == "Capomulin"]['Tumor Volume (mm3)']
```

```python
# Calculate the IQR and quantitatively determine if there are any potential outliers. 
quartiles = capomulin.quantile([.25,.5,.75])
lowerq = quartiles[0.25]
upperq = quartiles[0.75]
iqr = upperq-lowerq

print(f"The lower quartile of occupancy is: {np.round(lowerq,2)}")
print(f"The upper quartile of occupancy is: {np.round(upperq,2)}")
print(f"The interquartile range of occupancy is: {np.round(iqr,2)}")
print(f"The the median of occupancy is: {np.round((quartiles[0.5]),2)} ")

lower_bound = lowerq - (1.5*iqr)
upper_bound = upperq + (1.5*iqr)
print(f"Values below {np.round(lower_bound,3)} could be outliers.")
print(f"Values above {np.round(upper_bound,3)} could be outliers.")
```

```python
# Calculate the final tumor volume of each mouse in Ramicane treatment regime.
ramicane = merged_data2.loc[merged_data2["Drug Regimen"] == "Ramicane"]['Tumor Volume (mm3)']
```

```python
# Calculate the IQR and quantitatively determine if there are any potential outliers.
quartiles2 = ramicane.quantile([.25,.5,.75])
lowerq2 = quartiles2[0.25]
upperq2 = quartiles2[0.75]
iqr2 = upperq2-lowerq2

print(f"The lower quartile of occupancy is: {np.round(lowerq2,2)}")
print(f"The upper quartile of occupancy is: {np.round(upperq2,2)}")
print(f"The interquartile range of occupancy is: {np.round(iqr2,2)}")
print(f"The the median of occupancy is: {np.round((quartiles2[0.5]),2)} ")

lower_bound2 = lowerq2 - (1.5*iqr2)
upper_bound2 = upperq2 + (1.5*iqr2)
print(f"Values below {np.round(lower_bound2,3)} could be outliers.")
print(f"Values above {np.round(upper_bound2,3)} could be outliers.")
```

```python
# Calculate the final tumor volume of each mouse in Infubinol treatment regime.
infubinol = merged_data2.loc[merged_data2["Drug Regimen"] == "Infubinol"]['Tumor Volume (mm3)']
# Calculate the IQR and quantitatively determine if there are any potential outliers. 
quartiles3 = infubinol.quantile([.25,.5,.75])
lowerq3 = quartiles3[0.25]
upperq3 = quartiles3[0.75]
iqr3 = upperq3-lowerq3

print(f"The lower quartile of occupancy is: {np.round(lowerq3,2)}")
print(f"The upper quartile of occupancy is: {np.round(upperq3,2)}")
print(f"The interquartile range of occupancy is: {np.round(iqr3,2)}")
print(f"The the median of occupancy is: {np.round((quartiles3[0.5]),2)} ")

lower_bound3 = lowerq3 - (1.5*iqr3)
upper_bound3 = upperq3 + (1.5*iqr3)
print(f"Values below {np.round(lower_bound3,3)} could be outliers.")
print(f"Values above {np.round(upper_bound3,3)} could be outliers.")
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and linear regression model 
correlation = st.pearsonr(capomulin_weight,average_tumor_volume)
print(f"The correlation between both factors is {round(correlation[0],2)}")
```

```python
# Calculate the correlation coefficient and linear regression model 
# for mouse weight and average tumor volume for the Capomulin regimen
x_values = capomulin_weight
y_values = average_tumor_volume
(slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)
regress_values = x_values * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
plt.scatter(x_values,y_values)
plt.plot(x_values,regress_values,"r-")
plt.xlabel('Weight (g)')
plt.ylabel('Average Tumor Volume (mm3)')
plt.show()
```
